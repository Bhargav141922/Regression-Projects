#### BACKWARD ELIMINATION

```python
cols = list(x.columns)
pmax = 1
while (len(cols)>0):
    p=[]
    xc = x[cols]
    xc = sm.add_constant(xc)
    model = sm.OLS(y,xc).fit()
    p = pd.Series(model.pvalues.values[1:],index = cols)
    pmax = max(p)
    feature_with_pmax = p.idxmax()
    if(pmax > 0.05):
        cols.remove(feature_with_pmax)
    else:
        break
selected_features_BE = cols
print(selected_features_BE)
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
LR = LinearRegression()
RD = Ridge(normalize=True)
LS = Lasso(normalize=True)
EN = ElasticNet(normalize=True)
```

```python
from sklearn.model_selection import GridSearchCV
param_rd = {'alpha':np.arange(0.01,1,0.01)}
param_ls = {'alpha':np.arange(0.01,1,0.01)}
param_en = {'alpha':np.arange(0.01,1,0.01),'l1_ratio':np.arange(0.01,1,0.01)}
```

```python
GS_RD = GridSearchCV(RD,param_rd,cv=5,scoring='r2')
GS_RD.fit(x_final,y)
```

```python
GS_EN = GridSearchCV(EN,param_en,cv=5,scoring='r2')
GS_EN.fit(x_final,y)
```

```python
GS_EN.best_params_
```

```python
m2 = Ridge(alpha=0.35,normalize=True)
m3 = Lasso(alpha=0.99,normalize=True)
m4 = ElasticNet(alpha=0.23,l1_ratio=0.99,normalize=True)
```

```python
from sklearn.model_selection import KFold
from sklearn import metrics
kf=KFold(n_splits=5,shuffle=True,random_state=0)
for model,name in zip([LR,m2,m3,m4],['Linear_Regression','Ridge Regression','Lasso Regression','ElasticNet']):
    rmse=[]
    for train_idx,test_idx in kf.split(x_final,y):
        xtrain,xtest=x_final.iloc[train_idx,:],x_final.iloc[test_idx,:]
        ytrain,ytest=y.iloc[train_idx],y.iloc[test_idx]
        model.fit(xtrain,ytrain)
        y_pred=model.predict(xtest)
        mse=metrics.mean_squared_error(ytest,y_pred)
        rmse.append(np.sqrt(mse))
    print('RMSE scores: %0.03f (+/- %0.5f)[%s]' % (np.mean(rmse),np.std(rmse,ddof=1),name))
```

#### NOTE:

###### The drawback of GridSearch is that it only takes Bias Error(RMSE) into account and not the Variance Error.
